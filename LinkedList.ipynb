```python
class Node:
    
    def __init__(self,data):
        self.data = data 
        self.next = None

class LinkedList:

    def __init__(self):
        self.head = None
    
    # for displaying the node

    def display(self):
        temp = self.head
        while temp!=None:
            print(temp.data , end=' -> ')
            temp = temp.next
        print('None')
    
    # for appending a new element at the end  

    def append(self):
        temp = self.head
        while temp.next!=None:
            temp = temp.next
        temp.next = Node(input('Enter a new node: '))
    
    # for counting the number of nodes in the linked list 

    def count(self):

        temp = self.head
        count = 0 
        while temp!=None:
            count+=1
            temp = temp.next 
        print(f'Number of Nodes {count}')
    
    # for searching an element in the linked list 
    def search(self):

        temp = self.head
        Position = 0
        target = input('Enter the target data: ').strip()
        Found = False # Flag
        while temp!=None:
            Position +=1
            if temp.data == target:
                print(f'Target found at node {Position}')
                Found = True # Flag Set to True  
                break
            temp = temp.next
        # If the flag is still False
        if not Found:
            print(f'Sorry target data {target} not found!')
        
    # sorting the linked list using selection sort 

    def sort(self):

        swapped = True
        while swapped:
            swapped = False
            current = self.head
            while current.next != None:
                if current.data > current.next.data: 
                    current.data , current.next.data = current.next.data , current.data
                    swapped = True
                current = current.next 
    
    # for inserting an element at any position

    def insert(self):
        
        temp = self.head
        target = input('Enter a data to search -> ').strip()
        while temp!=None:
            if target == temp.data:

                # New node to be created
                NewNode = Node(input('New Node data : ')) 
                # assigning the address of the next node to the new node 
                NewNode.next = temp.next
                # assigning the address of the new node to previous node
                temp.next = NewNode
                print('Inserted!')
                break

            temp = temp.next

        if temp == None:
            print('Target not found!')

    # for removing an element any position

    def delete(self):

        temp = self.head
        prev = None
    
        target = input('Enter the data to be removed: ').strip()
        while temp!=None:
            if temp.next.data == target:
                prev = temp
                prev.next = temp.next.next
                break 
            temp = temp.next

    # Reverse the linked list 

    def reverse(self):
        
        temp = self.head
        while temp!=None:
            temp = temp.next

if __name__ == '__main__':

    Sll = LinkedList()
    while True:
        user = int(input('Choose anything from the given option \n 1) Create Node'))
        if user == 0:
            break
        elif user == 1:
            node = Node(input('New node data: '))
            Sll.head = node
        elif user == 2:
            Sll.display()
        elif user == 3:
            Sll.append()
        elif user == 4:
            Sll.count()
        elif user == 5:
            Sll.search()
        elif user == 6:
            Sll.sort()
        elif user==7:
            Sll.insert()
        elif user == 8:
            Sll.delete()
        elif user == 9:
            Sll.reverse()
        
```
